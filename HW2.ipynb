## Задание 1

Необходимо выделить все биграммы в этом тексте и сохранить их список в какую-либо переменную.

```python
a = input('Введите текст: \n') 
bigrams = [] 
text = a.split()
for i in range(len(text)-1): #проходимся циклом по всему тексту - 1 (так как мы добавляем сразу 2 элемента с индексом i и i+1, то есть следующий) 
     bigrams.append([text[i], text[i+1]] ) #добавляем в список наши биграммы
print(bigrams)
    
```

```python
#второй вариант , как в третьем задании 

b = input('Введите текст: \n') 
text1 = b.split() 
print([(text1[i], text1[i+1]) for i in range(len(text1)-1)])
```

Пройтись циклом по рабочей директории, читая содержимое каждого файла и записывая его в новый файл poems.txt По завершении цикла должен получиться файл со всеми стихотворениями!


```python
#чтобы работать в нужной папке - в нее необходимо перейти:
os.chdir('./poems')
print(os.getcwd()) #проверяем, что мы там, где нужно 
```

```python
for root, dirs, files in os.walk('.'):
    for name in files:
        with open(os.path.join(root, name), 'r', encoding='utf-8', errors='ignore') as f: #открываем каждый файл для чтения , используя модуль + игнорируем ошибки , которые видимо возникли из-за файла DS , прописываем в коде
            lyrics = f.read()   #читаем файл
            with open('poems.txt', 'a', encoding='utf-8', errors='ignore') as d: #в файл, который мы создали, дописываем информацию; режим доступа a, который позволяет это сделать 
                d.write(lyrics + '\n')
```

## Задание 3

Скопировать текст, сохранить его в переменную.
Узнать длину текста в символах.
Очистить текст от пунктуации, используя list compehensions. На выходе должен получиться список слов.

```python
a = """
Удивительная кошка

Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!"""
print(len(a))
from string import punctuation
poem = [i.strip(punctuation) for i in a.split()] #очищаем каждый элемент в списке от пунктуации, используя модуль. генератором списков проходимся по каждому элементу в уже разделенному по пробелам списке.  
print(poem)
```

Объединить слова из получившегося списка в строку через пробел.

```python
print(" ".join(poem)) #объединяем через пробел 
```

С помощью list comprehensions проверить, написано ли слово с большой буквы, и если да, то привести его к нижнему регистру. На выходе должен получиться список из всех слов стихотворения, написанных с маленькой буквы (не только те, которые изначально были с маленькой буквы, а все).

```python
print([w.lower() if w.istitle() else w for w in poem]) #приводим к нижнему регистру каждый элемент, который начинается с заглавной буквы, а если нет - то оставляем его
```
